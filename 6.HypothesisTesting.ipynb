## Hypothesis tests on sample datasets to compare population means 
### Objectives: 
1. Understand the concept of hypotgheesis testing. Formulate null and alternative hypotheses
2. Identify appropriate tests (Z test or t- tests). Perform hypothesis testing using python
3. Interpret test statistics and p values. Draw conclusions based on statistical evidence.

```python
#Given data 
pop_mean=100
pop_sd=15
visitors=[117, 119, 132, 106,106,133,121,102,118,103,103,113,81,84,101,94,114,
          96,88,131,106,111,88,101,111,92,115,100,105,100,137,109,94,122,91,113,80,90,112]
```

```python
#calculate sample mean and size
sample_mean = np.mean (visitors)
sample_size=len(visitors)
```

```python
#calculate z _score
z_score_calculated = (sample_mean - pop_mean) / (pop_sd / np.sqrt(sample_size))
```

```python
#calculate sample mean and size
sample_mean=np.mean(visitors)
sample_size=len(visitors)

#calculate Z-Score 
z_score_calculated=(sample_mean - pop_mean) / (pop_sd/np.sqrt(sample_size))

# FInd the criyical Z- value for two tailed test at alpha= 0.05 using scipy
alpha=0.05
z_critical=stats.norm.pdf(1- alpha/2)

#print results
print(f"Calulated Z- Score: {z_score_calculated:.3f}")
print(f"Calulated Z- Score: \u00b1 {z_critical:.3f}") #\u00b1: +-

#compare and conclude
if z_score_calculated > z_critical or z_score_calculated < -z_critical:
    print("Conclusion: Reject the null hypothesis. The sample mean is significantly different from the population mean. ")
else:
    print("Conclusion: Fail Reject the null hypothesis. The sample mean is not significantly different from the population mean.")
```

## Decision Rule: 
* Reject the Null Hypothesis (H<sub>0</sub>):
  1. If (Z<sub>calculated</sub>) lies outside the range[ (-Z<sub>calculated</sub>), Z<sub>calculated</sub>) ] , the sample mean is significantly different from the population mean.
  2. This means there is sufficient evidence to conclude that the observed difference is unlikely due to random sampling alone
* Fail to reject the null Hypothesis (H<sub>0</sub>1):
  1. If Z<sub>calculated</sub>) lies within the range [ (-Z<sub>calculated</sub>), Z<sub>calculated</sub>) ] , the sample mean is not significantly different from the population mean.
  2. This means the observed difference could reasonbly be due to random sampling.

## Use a Z-Test when all of the following conditions are satisfied: 
1. population standard deviation (σ) is known.
2. Sample size is large (usually n>=30)
3. Data is normally distributed or sample size is large enough for CLT
## Use a T-Test when any of the following is true: 
1. Population standard deviation (σ) is known.
2. Sample size is small (usually n< 30).A

# T-Test
### To test whether the average daily study time of students
### is significantly different  
### from the expected standard of 4 hours per daily

```python
# Daily study hours (in hours) reported by 10 students
study_hours=[3.5, 4.2, 3.8, 4.6, 5.0, 3.9, 4.1, 4.8, 3.7,4.3]
expected_mean=np.mean(study_hours)
sample_mean=np.mean(study_hours)
sample_std=np.std(study_hours, ddof=1)
n=len(study_hours)
print(" Sample Mean:", sample_mean)
print(" Sample Standard Deviation:", sample_std)
print(" Sample_size: ", n)
```

```python
t_stat, p_value = stats.ttest_1samp(study_hours, expected_mean)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

```python
alpha=0.05
if p_value < alpha:
   print("Conclusion: Reject the null hypothesis. Study time is significantly different from 4 hours.")
else:
    print("Conclusion: Fail Reject the null hypothesis.Study time is not significantly different from 4 hours.")
```
