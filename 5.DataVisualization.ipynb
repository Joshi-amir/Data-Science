# Theory:
What is data visualization?
Data Visualization is the graphical representation of data and information using visual elements such as charts, graphs, maps and dashboards.

Purpsoe of Data Visualization:
1. To make data easier to understand
2. To identify trends, patterns and outliers
3. To communicate insights effectively

Relation Between Data Analysis and Data Visualization 
1. Data analysis discovers what the data mmeans.
2. Data Visualization shows how to interprest those meanings visually
3. Together, they improve clarity, decision making, and communication.

Problem Statement 
A retail supermarket chain wants to evalute its monthly sales performance to understand customer behaviour, seasonal trends, and the influesnce of discounts on rvenue. As a data analyst, you are required to visualize the dataset and extract inshights that can assist management in planning promotions, inventory and sales strategies. 

Tools and Software Required
* Python 3.x
* Jupyter Notebook/ Google studio

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
data={
    "Month":["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
    "Sales": [120000,135000,150000,145000,160000,170000,180000,175000,165000,190000,210000,250000],
    "Customers": [420,450,470,460,480,500,520,510,490,530,560,600],
    "Discount": [5,5,10,8,10,12,15,12,10,8,5,5]
}
df=pd.DataFrame(data)
print(df)
```

# LINE CHART 

```python
plt.figure(figsize=(8,5))
plt.plot(df["Month"],df["Sales"],marker='o')
plt.title("Monthly sales trend")
plt.xlabel("Month")
plt.ylabel("Sales(NPR)")
plt.grid(True)
plt.show()
```

# BARCHART

```python
plt.figure(figsize=(8,5))
plt.bar(df["Month"],df["Sales"])
plt.title("Monthly sales trend")
plt.xlabel("Month")
plt.ylabel("Sales(NPR)")
plt.show()
```

# HEAT MAP(CORRELATION ANALYSIS)

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[["Sales", "Customers","Discount"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heat Map")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["Sales"], color="green")
plt.title("Sales DIstribution and outlier Detection")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df["Sales"], color="green")
plt.title("Sales DIstribution and outlier Detection")
plt.show()
```
