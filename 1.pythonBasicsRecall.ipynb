```python
max_score=max(scores.values())
min_score=min(scores.values())
highest_students=[name for name,score in scores.items() if score == max_score ]
lowest_students=[name for name,score in scores.items()if score==min_score]
print("Highest score:",max_score,"by",highest_students)
print("Lowest score:",min_score,"by",lowest_students)
```
