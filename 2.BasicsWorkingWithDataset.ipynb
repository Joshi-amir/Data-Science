```python
import pandas as pd 
dic={
    'name':["Elon","bill","sundar"],
    'age':[35,60,40],
    'city':['ktm','bkt','btd'],
    'salary':[1000,1000,1000]
}
print(dic['name']) 
df2=pd.DataFrame(dic)
print(df2[df2['salary']>80000])
df2['bonus']=df2['salary']*0.1
print(df2)
print(df2.describe())
print(df2.sort_values(by='age'))
df2.to_csv("output.csv",index=False)
```
