### Introduction: (Central Limit Theorem (CIT)): 
Regardless of the original distribution of the data, the distribution of the sample mean tends to be normal is the sample size is sufficientyly large.

This is one of the most imp ideas in data science, because it explains why normal distributions appear everywhere, even when data istself iis not normal.

```python
population = np.random.exponential (scale=2,size=100000)
plt.hist(population,bins=50,density=True,color='black')
plt.title("Population distibution(exponential-skewed)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
print(population.mean())
```

```python
np.random.seed(42)
population_size=100000
sample_size=[2,5,10,30,50,200] #sample sizes to demonstrate convergence
num_samples=1000 #number of samples for each sample size
population=np.random.exponential(scale=2,size=population_size)
print(population[:10]) #first 10 values
print(len(population)) ##size of population
population_mean=np.mean(population)
print(population_mean)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(population, bins=50, kde=True, color='darkgreen', alpha=0.95) #kde=kernal density 
plt.title("Population Distribution (Exponential)", fontsize=17)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(14,10))
for i, size in enumerate(sample_size, start=1):
    sample_means=[np.mean(np.random.choice(population, size=size, replace=False))
                  for _ in range(num_samples)]
    #print(sample_means)
    #plot the distribution of sample means
    plt.subplot(3,2,i)
    sns.histplot(sample_means, bins=30, kde=True, color='skyblue', alpha=0.9)
    plt.title(f"Sample means distribution (n={size})", fontsize=14)
    plt.xlabel("Sample Mean", fontsize=12)
    plt.ylabel("Frequency", fontsize=12)
#Showing the final plots
plt.suptitle("Demonstration of the central Limit Theorem", fontsize=18, y=1.02)
plt.tight_layout()
plt.show()
```
